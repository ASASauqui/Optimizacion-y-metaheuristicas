## ==================== Libraries ====================

```python
# Determinadas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import random
from sklearn.metrics import mean_squared_error, f1_score
from sklearn.metrics import accuracy_score
from image_similarity_measures.quality_metrics import rmse, ssim, sre

# Procesamiento de imágenes
from PIL import Image
import cv2
```

### Global Variables

```python
cont_call_fo = 0
```

## <br><br><br>==================== Functions ====================

### Objective Function

```python
def objectiveFunction(objective_img, transformed_img):
    # Aumentar contador.
    global cont_call_fo
    cont_call_fo +=1
    
    # Obtener la forma de la image.
    h, w = objective_img.shape
    
    # Total de pixeles.
    total = h*w
    
    # Sacar f1.
    score = f1_score(objective_img, transformed_img, average = 'micro')*total
    
    # Restarle al score la cantidad de no ceros que hay de la imagen resultante entre ambas imágenes.
    score -= np.count_nonzero(objective_img-transformed_img)
        
    return score  
```

### Create population

```python
def createPopulation(n, w, h, epsilon=0.001):
    # Matriz de población.
    population = np.zeros((n,8))
    
    # Primera parte.
    population[:, 0] = np.random.uniform(-w/2,w/2,n)
    population[:, 1] = np.random.uniform(-h/2,h/2,n)
    population[:, 2] = np.random.uniform(-180,180,n)
    population[:, 3] = np.random.uniform(0.5,1.5,n)
    
    # Segunda parte.
    population[:, 4] = np.abs( np.random.normal(loc=0,scale=w/10,size=n) )
    population[np.where(population[:, 4] < epsilon), 4] = epsilon
    
    population[:, 5] = np.abs( np.random.normal(loc=0,scale=h/10,size=n) )
    population[np.where(population[:, 5] < epsilon), 5] = epsilon
    
    population[:, 6] = np.abs( np.random.normal(loc=0,scale=30,size=n) )
    population[np.where(population[:, 6] < epsilon), 6] = epsilon
    
    population[:, 7] = np.abs( np.random.normal(loc=0,scale=0.1,size=n) )
    population[np.where(population[:, 7] < epsilon), 7] = epsilon
    
    return population
```

### Create transformed image

```python
def getImageTransformation(img, individual):    
    # Obtener la forma de la imagen.
    h, w = img.shape
    
    # Copiar imagen.
    transformed_img = img.copy()
     
    # Creación de la matriz de escala y traslación.
    matrix = np.float32([[individual[3], 0, individual[0]], 
                        [0, individual[3], individual[1]]])
    # Tranformar.
    transformed_img = cv2.warpAffine(transformed_img.copy(), matrix, (w, h))
    
    # Creación de la matriz de rotación.
    matrix = cv2.getRotationMatrix2D((w/2, h/2), individual[2], 1)
    # Tranformar.
    transformed_img = cv2.warpAffine(transformed_img.copy(), matrix, (w, h))
    
    return transformed_img
```

### Recombination function

```python
def intermediateRecombination(parentA, parentB):
    return (parentA + parentB)/2
```

```python
def recombination(mu, lambda_offspring, population):
    # Cantidad de offsprings.
    offspring_size = int(mu*lambda_offspring)
    
    # Matriz de offsprings.
    offspring = np.zeros((offspring_size, 8))
    
    for i in range(offspring_size):
        # Elegir dos papás random.
        random_numbers = np.random.choice(range(0,mu), 2, replace=False)
        
        # Obtener un offspring.
        offspring[i] = intermediateRecombination(population[random_numbers[0]], population[random_numbers[1]])
        
    return offspring
```

### Mutation function

```python
def mutation(population, epsilon=0.001):
    # Tamaño de los individuos.
    N = len(population)
    
    # Tau.
    tau = 1/math.sqrt(4)
    
    for i in range(N):   
        # Primera parte del offspring.
        population[i,0] += np.random.normal(loc=0,scale=population[i,4])
        population[i,1] += np.random.normal(loc=0,scale=population[i,5])
        population[i,2] += np.random.normal(loc=0,scale=population[i,6])
        population[i,3] += np.random.normal(loc=0,scale=population[i,7])

        # Segunda parte del offspring.
        population[i,4] = population[i,4] * np.exp( np.random.normal(loc=0,scale=tau) )
        population[i,5] = population[i,5] * np.exp( np.random.normal(loc=0,scale=tau) )
        population[i,6] = population[i,6] * np.exp( np.random.normal(loc=0,scale=tau) )
        population[i,7] = population[i,7] * np.exp( np.random.normal(loc=0,scale=tau) )
        
    return population
```

### Survivor functions

```python
def getBestSurvivors(survivors_fitness, mu):
    return np.argsort(survivors_fitness)[::-1][:mu]
```

```python
def survivor(offspring, parents, offspring_fitness, parents_fitness, mu, mode):
    # Índices de los supervivientes.
    survivor_indexes = np.zeros(mu)
    
    if mode == "BEST_OFFSPRING":
        # Obtener mejores supervivientes.
        survivor_indexes = getBestSurvivors(offspring_fitness, mu)
        
        return offspring[survivor_indexes], offspring_fitness[survivor_indexes]
    elif mode == "BEST_PARENTS_PLUS_OFFSPRING":
        # Combinación de individuos.
        individual_combination = np.concatenate((offspring, parents), axis=0)
        
        # Combinación de fitness.
        fitness_combination = np.concatenate((offspring_fitness, parents_fitness))
        
        # Obtener mejores supervivientes.
        survivor_indexes = getBestSurvivors(fitness_combination, mu)
        
        return individual_combination[survivor_indexes], fitness_combination[survivor_indexes]
    
    return None,None
```

### Elite function

```python
def getElite(fitness):
    # Obtener el tamaño de la población.
    N = len(fitness)
    
    # Obtener el índice del mejor.
    index = np.argmax(fitness)
            
    return index
```

### Evolution algorithm

```python
def evolutionStrategies(mu, lambda_offspring, G, start_img, objective_img):
    # Obtener la forma de la imagen.
    h, w = objective_img.shape
    
    # Crear población inicial.
    population = createPopulation(mu, w, h)
    
    # Arreglo de fitness de la población.
    population_fitness = np.zeros(mu)
    offspring_fitness = np.zeros(int(mu*lambda_offspring))
    
    # Calcular el fitness de la población.
    for i in range(mu):
        # Obtener la imagen transformada del individuo.
        transformed_img = getImageTransformation(start_img, population[i])
        
        # Calcular el fitness del individuo.
        population_fitness[i] = objectiveFunction(objective_img, transformed_img)
        
    # Obtener el elite.
    eliteIndex = getElite(population_fitness)
    elite = [ population[eliteIndex], population_fitness[eliteIndex] ]
    
    # Elite previo.
    prevElite = elite
    
    # Impresión inicial.
    transformed_img = getImageTransformation(start_img.copy(), prevElite[0])
    result = objectiveFunction(objective_img, transformed_img.copy())
    plt.figure()
    plt.imshow(transformed_img)
    plt.title("Gen: Inicio, Mejora: " + str(elite[1]) + ", Individuo: " + str(elite[0]) )
    plt.show()
        
    # Generaciones.
    no_iter = 0
    while no_iter < G:
        # Recombinación.
        offspring = recombination(mu, lambda_offspring, population.copy())
        
        # Mutación.
        offspring = mutation(offspring.copy())
        population = mutation(population.copy())
        
        # Calcular el fitness de la población.
        for i in range(mu):
            # Obtener la imagen transformada del offspring.
            transformed_img = getImageTransformation(start_img, population[i])

            # Calcular el fitness del offspring.
            population_fitness[i] = objectiveFunction(objective_img, transformed_img)
        
        # Calcular el fitness del offspring.
        for i in range(int(mu*lambda_offspring)):
            # Obtener la imagen transformada del offspring.
            transformed_img = getImageTransformation(start_img, offspring[i])

            # Calcular el fitness del offspring.
            offspring_fitness[i] = objectiveFunction(objective_img, transformed_img.copy())

        # Survivor.
        population, population_fitness  = survivor(offspring.copy(), population.copy(), offspring_fitness.copy(), population_fitness.copy(), mu, "BEST_OFFSPRING")
        
        # Obtener el elite.
        elite = [ population[0], population_fitness[0] ]
        
        # Poner el elite en la población.
        population[mu-1] = elite[0]
        population_fitness[mu-1] = elite[1]
        
        # Si el elite actual es mejor que el global.
        if elite[1] > prevElite[1]:
            # Cambiar al mejor elite global.
            prevElite = elite
            
            # Mostrar resultados.
            transformed_img = getImageTransformation(start_img.copy(), elite[0])
            result = objectiveFunction(objective_img, transformed_img.copy())
            plt.figure()
            plt.imshow(transformed_img)
            plt.title("Gen:" + str(no_iter) + ", Mejora: " + str(elite[1]) + ", Individuo: " +  str(elite[0]) )
            plt.show()
        
        # Aumentar iterador.
        no_iter += 1
        
    return prevElite
```

### Load dataset

```python
# Direcciones del dataset.
path1 = r'Datasets/panamericana1.PNG'
path2 = r'Datasets/panamericana2.PNG'

# Leer documento .csv.
start_img = cv2.imread(path1)
objective_img = cv2.imread(path2)

# Imagen a escala de grises.
start_img = cv2.cvtColor(start_img.copy(), cv2.COLOR_BGR2GRAY)
objective_img = cv2.cvtColor(objective_img.copy(), cv2.COLOR_BGR2GRAY)

# # Binarización.
start_img = cv2.threshold(start_img.copy(), 200, 255, cv2.THRESH_BINARY_INV)[1]
start_img[start_img > 0] = 1
objective_img = cv2.threshold(objective_img.copy(), 200, 255, cv2.THRESH_BINARY_INV)[1]
objective_img[objective_img > 0] = 1

# Mostrar imagen.
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.imshow(start_img, cmap="gray")
plt.title("Start Image")
plt.subplot(1,2,2)
plt.imshow(objective_img, cmap="gray")
plt.title("Objective Image")
plt.show()
```
