```python
def objectiveFunction(P, ShF2, ShM1): # Shaking force, Shaking moment.
    # Descomponer en variables.
    x_c1,x_c2,x_c3,y_c1,y_c2,y_c3,t_c1,t_c2,t_c3 = P
    
    # Obtener el resultado de las funciones.
    resShF2 = ShF2(x_c1,x_c2,x_c3,y_c1,y_c2,y_c3,t_c1,t_c2,t_c3)
    resShM1 = ShM1(x_c1,x_c2,x_c3,y_c1,y_c2,y_c3,t_c1,t_c2,t_c3)
    
    # Puntos.
    A = np.array((0, 0))
    B = np.array((resShF2, resShM1))
    
    # Distancia euclidiana.
    dist = np.sqrt(np.sum(np.square(A-B)))
    
    return dist
```

```python
# Variables a utilizar.
no_points = 100
points = np.zeros((no_points,2))
fitness = np.zeros(no_points)

# Mejores resultados.
best = None
best_fit = float('inf')


for i in range(no_points):
    # Llamar algoritmo evolutivo diferencial.
    res = differential_algorithm(objectiveFunction, bounds, args=(ShF2, ShM1), popsize=30, ngen=500, cr=0.5)
    
    # Añadir fitness.
    fitness[i] = res['fun']
    
    # Obtener punto.
    points[i] = getPoints(res['P'], ShF2, ShM1)
    
    # Si el resultado mejora al global.
    if fitness[i] < best_fit:
        best_fit = fitness[i]
        best = res
    
    # Plotear puntos.
    plt.figure()
    plt.scatter(points[:i+1,0], points[:i+1,1], c=fitness[:i+1])
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xlabel("ShF")
    plt.ylabel("ShM")
    plt.title("Point No.: " + str(i+1) + ", Point: " + str(points[i]) + ", Fitness = " + str(res['fun']))
    plt.show()
```

```python
# El menor resultado.
min_arg = np.argmin(fitness)

# Mostrar resultado.
print("El mejor fue:\nShF = " + str(points[min_arg][0]) + 
      "\nShM = " + str(points[min_arg][1]) + 
      "\nFitness de: " + str(best['fun']), '\n')
vector = best['P']
for i in range(3):
    print("x" + str(i+1) + ", y" + str(i+1) + ", t" + str(i+1) + " =", vector[i], vector[i+3], vector[i+6])
```
